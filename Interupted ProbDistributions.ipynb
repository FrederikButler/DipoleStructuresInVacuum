```python
import numpy as np
import matplotlib.pyplot as plt
import Qchains
import plotly.graph_objects as go
from scipy.optimize import curve_fit

from matplotlib.colors import LinearSegmentedColormap, Normalize
import matplotlib.cm as cm
from matplotlib.colors import hsv_to_rgb
import cmath
```

```python
points_10 = Qchains.ArrayGenerator(50)
dipole_vector_hat_10 = np.array([0, 1, 0]) #i y retning 
d_10 = 0.5
eig_val_10, eig_vec_10 = Qchains.FindEigenstates(points_10, dipole_vector_hat_10, d_10)
Qchains.SiteProbAndAmplitudeDistribution(eig_val_10, eig_vec_10, 0, titleprecision = 8, sharey=False)
```

```python
chain_40 = Qchains.ArrayGenerator(40)
chain41 = Qchains.ArrayGenerator(41)
middle_index_41 = len(chain41) // 2
chain_40c_missing = np.delete(chain41, middle_index_41, axis=0)

dipole_vector_hat_40 = np.array([0, 1, 0]) #i y retning 
d_40 = 0.5

eig_val_40_interrupted, eig_vec_40_interrupted = Qchains.FindEigenstates(chain_40c_missing, dipole_vector_hat_40, d_40)
Qchains.SiteProbAndAmplitudeDistribution(eig_val_40_interrupted, eig_vec_40_interrupted, 0, titleprecision = 7, sharey=False, interrupted_array=True)

eigval_40, eigvec_40 = Qchains.FindEigenstates(chain_40, dipole_vector_hat_40, d_40)
Qchains.SiteProbAndAmplitudeDistribution(eigval_40, eigvec_40, 0, titleprecision = 7, sharey=False, interrupted_array=False)
```

```python
for i in range(40):
    Qchains.SiteProbAndAmplitudeDistribution(eig_val_40_interrupted, eig_vec_40_interrupted, i, titleprecision = 7, sharey=False, interrupted_array=True)
```

```python
for i in range(40):    
    Qchains.SiteProbAndAmplitudeDistribution(eigval_40, eigvec_40, i, titleprecision = 7, sharey=False, interrupted_array=False)
```
